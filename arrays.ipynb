```python
#sum of array
arr=[1,2,3,4,5]
sum=0
for i in range(len(arr)):
    sum+=arr[i]
print("Sum of the elements of the array :",sum)
```

```python
#Reverse an array
arr=[1,2,3,4,5,6]
i=0
j=len(arr)-1
while i<=j:
    arr[i],arr[j]=arr[j],arr[i]
    i+=1
    j-=1
print(arr)
```

```python
#largest element of an array
arr=[90,78,34,12,45,92]
mx=arr[0]
for i in range(1,len(arr)):
    if mx<arr[i]:
        mx=arr[i]
print("Largest element of the array is",mx)
```

```python
#Mean of array
arr=[23,45,2,18,20]
sum=0
n=len(arr)
for i in range(len(arr)):
    sum+=arr[i]
print("Mean of array:",sum//n)
```

```python
#second largest element of the array
arr=[34,92,40,12,90,5]
mx=arr[0]
smx=-1
for i in range(1,len(arr)):
    if mx<arr[i]:
        smx,mx=mx,arr[i]
    elif mx>arr[i] and smx<arr[i]:
        smx=arr[i]
print("Second largest element of the array:",smx)
```
